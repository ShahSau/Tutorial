```python
output_notebook()
# import the HoverTool
from bokeh.models import HoverTool

# Create the figure: p
p = figure(x_axis_label='fertility', y_axis_label='female_literacy (% population)')

# Add a circle glyph to the figure p
p.circle(fertility_latinamerica,female_literacy_latinamerica,size=10,hover_fill_color='firebrick', hover_alpha=0.5,
         hover_line_color='white'
)

# Add an x glyph to the figure p
p.x(fertility_africa,female_literacy_africa,size=10,color='red')

# Specify the name of the file
#output_file('fertlityvseducation_latin_africa.html')
# Create a HoverTool: hover
hover = HoverTool(tooltips=[('index', '$index'), ('(x,y)', '($x,$y)')], mode='vline')

# Add the hover tool to the figure p
p.add_tools(hover)

# Display the plot
show((p),notebook_handle=True)
```

```python
from bokeh.models import ColumnDataSource
from bokeh.plotting import figure, output_file, show

output_notebook()
#output_file("hbar_stack.html")

source = ColumnDataSource(data=dict(
    x=[1, 2, 3, 4, 5],
    y1=[1, 2, 4, 3, 4],
    y2=[1, 4, 2, 2, 3],
))
p = figure(plot_width=400, plot_height=400)

p.vbar_stack(['y1', 'y2'], x='x', width=0.8, color=("grey", "lightgrey"), source=source)

show((p),notebook_handle=True)
```

```python
# Import numpy as np
import numpy as np
output_notebook()
# Create array using np.linspace: x
x = np.linspace(0,5,100)

# Create array using np.cos: y
y = np.cos(x)

# Add circles at x and y
p = figure(plot_width=400, plot_height=400)
p.circle(x,y)

# Specify the name of the output file and show the result
output_file('numpy.html')
show((p),notebook_handle=True)
```

```python
#Import CategoricalColorMapper from bokeh.models
from bokeh.models import CategoricalColorMapper

# Convert df to a ColumnDataSource: source
source = ColumnDataSource(df)

# Make a CategoricalColorMapper object: color_mapper
color_mapper = CategoricalColorMapper(factors=['europe', 'japan', 'usa'],
                                      palette=['red', 'green', 'blue'])
p2=figure(x_axis_label='weight', y_axis_label='mile-per-gallon')
# Add a circle glyph to the figure p
p2.circle('weight', 'mpg', source=source,
            color={'field':'origin', 'transform':color_mapper},
            legend='origin')

# Specify the name of the output file and show the result
#output_file('colormap.html')
show((p2),notebook_handle=True)
```
